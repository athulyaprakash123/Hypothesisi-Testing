```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.distplot(data['Unit A'])
```

```python
sns.distplot(data['Unit B'])
```

null hypothesis = there is significant difference of cutlets between 2 units

alternate hypothesis = there is no significant difference of cutlets between 2 units

```python
stats.f_oneway(data['Unit A'],data['Unit B'])
```

we got pvalue as 0.472239 which is less than 0.5 so from our statement,since the value is less than 0.5 it is a alternate hypothesis therefore alternate hypothesis is accepted.

# 3. Costomer+OrderForm

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
data=pd.read_csv('Costomer+OrderForm.csv')
data
```

```python
data.head()
```

```python
#SKLEARN FROM STRING TO FLOAT
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Phillippines']=le.fit_transform(data['Phillippines'])
data['Indonesia']=le.fit_transform(data['Indonesia'])
data['Malta']=le.fit_transform(data['Malta'])
data['India']=le.fit_transform(data['India'])
```

```python
data
```

```python
data['Phillippines'].skew()
```

```python
data['Indonesia'].skew()
```

```python
data['Malta'].skew()
```

```python
data['India'].skew()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
stats.f_oneway(data['Phillippines'],data['Indonesia'],data['Malta'],data['India'])
```

null hypothesis = defective % vary by centre

alternate hypothesis = defective % dosen't vary by centre

since the pvalue is 0.27 which is less than 0.05 alternate hypothesis is accepted
